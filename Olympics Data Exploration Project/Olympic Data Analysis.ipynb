```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Upload dataset
athletes = pd.read_csv(r"C:\Users\moahm\Documents\Portfolio Project\Olympics Dataset\athlete_events.csv")
regions = pd.read_csv(r"C:\Users\moahm\Documents\Portfolio Project\Olympics Dataset\noc_regions.csv")
```

```python
# Show data
athletes.head()
```

```python
regions.head()
```

```python
# Join the dataframe
athletes_df = athletes.merge(regions , how = 'left', on='NOC')
athletes_df.head()
```

```python
#last 5 rows
athletes_df.tail()
```

```python
# Column names consistent
athletes_df.rename(columns = {'region':'Region','notes':'Notes'}, inplace=True)
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df['Sport'].value_counts()
```

```python
athletes_df['Medal'].value_counts()
```

```python
# Check if there is any missing values
missing_value = athletes_df.isna()
missing_columns = missing_value.any()
missing_columns
```

```python
# Let's count missing values
missing_value_count = athletes_df.isnull().sum()
missing_value_count
```

```python
# How many cities Olympics has been held in so far? 
Total_cities = len(athletes_df['City'].unique())#
Total_cities
```

```python
# Australia Details
athletes_df.query('Team =="Australia"').head(5)
```

```python
# Top countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for the top 10 countries
plt.figure(figsize=(12,6))
plt.title("Overall Participation by Country")
sns.barplot(x=top_10_countries.index, y =top_10_countries, palette='Set2');
```

```python
# What are the winter sports?
winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# What are the summer sports?
summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# Age Distribution of the participants
sns.set_style('darkgrid')
plt.figure(figsize=(12,6))
plt.title("Age distribution of the athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins=np.arange(10,80,2), color='seagreen');
```

```python
# male and female participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pie plot for male and female athletes
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=150, shadow=False)
```

```python
women_in_olympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=women_in_olympics, palette="Spectral")
plt.title('Women Participation')
```

```python
men_in_olympics = athletes_df[(athletes_df.Sex == 'M') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=men_in_olympics, palette="Spectral")
plt.title('Men Participation')
```

```python
part = women_in_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:, 'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
# Gold metal athletes
gold_Medals = athletes_df[(athletes_df.Medal == 'Gold')]
gold_Medals.head()
```

```python
# Gold medals from each regions
gold_Medals.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
totalGold_Medals = gold_Medals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="index", y="Medal", data=totalGold_Medals,
               height=5, kind="bar", palette="rocket" )

g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per Country');
```

```python
max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team
team_names.value_counts().head()
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
plt.ylabel(None);
plt.xlabel('Countrywise Medals for the year 2016');
```

```python
not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize=(12,10))
axis =sns.scatterplot(x='Height', y='Weight', data=not_null_medals, hue='Sex', s=150)
plt.title('Height vs Weight of Olympics Medalists')
```
